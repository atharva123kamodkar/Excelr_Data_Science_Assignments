```python
import pandas as pd
import mlxtend
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df=pd.read_csv(r"C:\DataScience Assignments\Association Rule\book.csv")
df
```

```python
df.value_counts()
```

Apriori Algorithm


1)Association Rule With 10% Support and 70% confidence

```python
frequent=apriori(df,min_support=0.1,use_colnames=True)
frequent
```

```python
rules=association_rules(frequent,metric='lift',min_threshold=0.7)
rules
```

```python
rules.sort_values('lift',ascending=False)
```

```python
import matplotlib.pyplot as plt
plt.scatter(rules['support'],rules['confidence'])
```

2)Association Rule with 20% support and 60% confidence

```python
frequent2=apriori(df,min_support=0.2,use_colnames=True)
frequent2
```

```python
#60% confidence
rules1=association_rules(frequent2,metric='lift',min_threshold=0.6)
rules1
```

```python
plt.scatter(rules1['support'],rules1['confidence'])
```
