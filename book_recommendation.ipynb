Cosine Similarity

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
user_sim=1-pairwise_distances(new_books.values,metric='cosine')
user_sim
```

```python
user_simdf=pd.DataFrame(user_sim)
user_simdf
```

```python
user_simdf.index=df.userid.unique()
user_simdf.columns=df.userid.unique()
```

```python
import numpy as np
np.fill_diagonal(user_sim,0)
user_simdf
```
