```python
#normalization
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return x
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='average'))
```

K-Means

```python
df2=pd.read_csv("C:\DataScience Assignments\Clustering\crime_data.csv")
```

```python
df2
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_df=scaler.fit_transform(df2.iloc[:,1:])
scaled_df
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,random_state=0)
    kmeans.fit(scaled_df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('No.Of Cluster')
plt.ylabel('wcss')
plt.show()
```

```python
kmeans=KMeans(n_clusters=4,random_state=0)
predict=kmeans.fit_predict(scaled_df)
```

```python
predict
```

```python
clusters_new=KMeans(n_clusters=4,random_state=0)
new_predict=clusters_new.fit_predict(scaled_df)
```
