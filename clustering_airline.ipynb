HierarchicalClustering

```python
def normfunc(i):
    x=(i-i.min())/(i.max()-i.min())
    return x
```

```python
df_norm=normfunc(df)
```

```python
df_norm
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='single'))
```

```python
hc=AgglomerativeClustering(n_clusters=4,metric='euclidean',linkage='single')
```

```python
hc
```

```python
y_hc=hc.fit_predict(df_norm)
y_hc
```

```python
clusters=pd.DataFrame(y_hc,columns=['clusterno'])
clusters
```

```python
df['h_clusterid']=clusters
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_df=scaler.fit_transform(df)
scaled_df
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmean=KMeans(n_clusters=i,init='k-means++',max_iter=300,random_state=0)
    kmean.fit(scaled_df)
    wcss.append(kmean.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('No of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmean=KMeans(n_clusters=4,random_state=0)
predict=kmean.fit_predict(scaled_df)
```

```python
clusters_new=KMeans(n_clusters=4,random_state=0)
new_predict=clusters_new.fit_predict(scaled_df)
```
